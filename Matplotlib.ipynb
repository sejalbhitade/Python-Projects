```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# line plot

```python
x=[1,3,5,7,9,11]
y=[11,14,15,19,24,26]
```

```python
plt.plot(x,y,marker="o")
plt.grid()
plt.show()
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
plt.plot(x,y,marker="*")   #marker is used to define the plot points
plt.grid()
plt.show()
```

```python
area=[600,300,400,1000,1200]
price=[15,34,28,40,35]
```

```python
plt.plot(area,price,marker="*")
plt.grid()
plt.show()
```

## figsize,title,labels

```python
plt.figure(figsize=(2,2))  #it is used to give the size of the outer box
plt.title("area v/s price")
plt.xlabel("area in sq.ft")
plt.ylabel("price in lakhs")
plt.plot(area,price,marker="*")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4,2))  #it is used to give the size of the outer box
plt.title("area v/s price")
plt.xlabel("area in sq.ft")
plt.ylabel("price in lakhs")

plt.xlim(400,1500)
plt.ylim(10,42)

plt.xticks(rotation=30)  #it rotates the axis to the specifies given degree
plt.yticks(rotation=30)

plt.plot(area,price,marker="*",color='purple')
plt.grid()
plt.show()
```

```python
x=np.arange(1,20)
y1=np.sqrt(x)
y2=np.log(x)

plt.plot(x,y1,color='blue',label='sqrt')
plt.plot(x,y2,color='orange',label='log')
plt.legend()  #it is used to define scale in the graph

plt.grid()
plt.show()
```
