```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cars.csv")
df.head()
```

## finding discriptive statistcs

```python
df.describe()
```

```python
#mean > median --------> right skewness
#most of the data is on the left side and few larger values are on the right
#median > mean left skewness
#most of the data is on the rioght side and few larger values are on the left
```

```python
plt.figure(figsize=(4,4))
plt.hist(df['width'],bins=10)
plt.title('width frequency distribution')
plt.xlabel('width')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.hist(df['price'],bins=10)
plt.title('price frequency distribution')
plt.xlabel('price')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.hist(df['height'],bins=10)
plt.title('height frequency distribution')
plt.xlabel('height')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

## plot mean and median

```python
plt_mean=df['price'].mean()
plt_median=df['price'].median()

plt.figure(figsize=(6,6))
plt.hist(df['price'],bins=10)

plt.axvline(plt_mean,color='red')
plt.axvline(plt_median,color='yellow')
            
plt.title('priceuency distribution')
plt.xlabel('price')
plt.ylabel('Frequency')
plt.grid()
plt.show()            
```

```python
plt_mean=df['width'].mean()
plt_median=df['width'].median()

plt.figure(figsize=(6,6))
plt.hist(df['width'],bins=10)

plt.axvline(plt_mean,color='red')
plt.axvline(plt_median,color='yellow')
            
plt.title('width ceuency distribution')
plt.xlabel('width')
plt.ylabel('Frequency')
plt.grid()
plt.show()            
```

```python
plt_mean=df['height'].mean()
plt_median=df['height'].median()

plt.figure(figsize=(6,6))
plt.hist(df['height'],bins=10,color='pink')

plt.axvline(plt_mean,color='green')
plt.axvline(plt_median,color='yellow')
            
plt.title('height frequency distribution')
plt.xlabel('height')
plt.ylabel('Frequency')
plt.grid(color='red')
plt.show()            
```

# Pie chart

#### it is a circular statistical graph

#### it shows percentage values

### 1. using pandas 

```python
df['body-style'].value_counts()
```

```python
#outopct - values in %
#explode - slices out parts in order to highlight them 
# f - gives floating values 

plt.figure(figsize=(6,6))
df['body-style'].value_counts().plot.pie(autopct="%1.2ff%%",explode=(0,0,0.3,0,0))
plt.show
```

## 2. using matplotlib

```python
plt.figure(figsize=(6,6))
x=list(df['body-style'].unique())
y=list(df['body-style'].value_counts())
plt.pie(y,autopct="%1.3f%%",labels=x)
plt.show()
```

# Scatter plot

```python
df.corr()
```

#### correlation close to 1 -----> positive correlation ------> 1 variable increases other increases

#### correlation close to -1------> negetive correlation ------> 1 variable increases other decreases

#### correlation value is 0.0 , 0.1 -----> zero correlation ----> no relation between variables.

```python
plt.figure(figsize=(6,6))
plt.scatter(x=df['width'],y=df['price'])
plt.title("width v/s price correlation")
plt.xlabel("width")
plt.ylabel("price")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x=df['symboling'],y=df['price'])
plt.title("symboling v/s price correlation")
plt.xlabel("symboloing")
plt.ylabel("price")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x=df['city-mpg'],y=df['price'])
plt.title("city-mpg v/s price correlation")
plt.xlabel("city-mpg")
plt.ylabel("price")
plt.show()
```
