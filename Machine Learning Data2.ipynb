```python
#replace the missing values with NaN

df['Python'].replace('missing',np.nan,inplace=True)
df['Machine Learning'].replace('missing',np.nan,inplace=True)
df['Machine Learning'].replace('?',np.nan,inplace=True)
df['Age'].replace('?',np.nan,inplace=True)
df['AI'].replace('missing',np.nan,inplace=True)
df['NLP'].replace('?',np.nan,inplace=True)
df['Cv'].replace('missing',np.nan,inplace=True)
df['Cv'].replace('?',np.nan,inplace=True)



```

```python
#change the datatype

df['Python']=df['Python'].astype(float)
df['Machine Learning']=df['Machine Learning'].astype(float)
df['Age']=df['Age'].astype(float)
df['AI']=df['AI'].astype(float)
df['NLP']=df['NLP'].astype(float)
df['Cv']=df['Cv'].astype(float)
```

```python
#get the mean

pm=df['Python'].mean()
mm=df['Machine Learning'].mean()
am=df['Age'].mean()
aam=df['AI'].mean()
nm=df['NLP'].mean()
cm=df['Cv'].mean()

```

```python
#fillna() fill the null values

df['Python'].fillna(pm,inplace=True)
df['Machine Learning'].fillna(mm,inplace=True)
df['Age'].fillna(am,inplace=True)
df['AI'].fillna(aam,inplace=True)
df['NLP'].fillna(nm,inplace=True)
df['Cv'].fillna(cm,inplace=True)
```
