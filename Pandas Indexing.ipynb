#### df.loc[rows,cols]

## col variations

```python
#display all the rows of A to C col

df.loc[:,'A':'C']
```

```python
#display records from b to d

df.loc[:,'B':'D']
```

```python
#Display all records from A to D

df.loc[:,['A','D']]
```

## Row variations 

```python
#display cols with a,c,f rows

df.loc[['a','c','f'],:]
```

```python
#display all cols with c to f rows

df.loc['c':'f',:]
```

```python
#display alternate rows of all cols

df.loc[::2]
```

```python
#display a to e rows of a to c cols 

df.loc['a':'e','A':'C']
```

```python
#Display alternate rows and alternate cols

df.loc[::2,::2]
```

# Integer - Based Indexing------->df.iloc()

```python
df
```

```python
#display all the rows of 1st 2 cols

df.iloc[:,0:2]
```

```python
#display 5 rows of 1st 2 cols

df.iloc[:5,:2]
```

```python
#display the last element

df.iloc[-1,-1]
```

```python
#display the reverse of the table

df.iloc[::-1,::-1]
```
