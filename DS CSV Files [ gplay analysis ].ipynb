```python
#copy paste the unnamed column as it is dnt write directly it will give error

df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

#### astype function - to change the datatype 

```python
df.isnull().sum()   #finding missing values
```

```python
df.isnull().sum()/len(df)*100   #finding missing value percent wise 
```

```python
#hence is more than 3% we need to do fillna()

rm=df['Rating'].mean()
df['Rating'].fillna(rm,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#to drop whole dataset ka values

df['Reviews']=df['Reviews'].astype(float)
```

```python
df.info()
```

```python
df['Size'].value_counts()
```

```python
def clean(x):
    if 'k' in x:
        x=x.replace('k','')
        return float(x)/1024
    elif 'M' in x:
        x=x.replace('M','')
        return float(x)
    elif 'Varies with device' in x:
        return np.nan
    else:
        return float(x)
```

```python
df['Size']=df['Size'].map(clean)
```

```python
df['Size'].value_counts()
```

```python
df.info()
```

```python
df['Installs'].value_counts()
```

```python
df['Installs']=df['Installs'].str.replace('+','')
df['Installs']=df['Installs'].str.replace(',','')
```

```python
df['Price'].value_counts()
```

```python
df['Installs']=df['Installs'].astype(float)
```

```python
df.info()
```

```python
df['Price']=df['Price'].str.replace('$','')
df['Price']=df['Price'].astype(float)
```
